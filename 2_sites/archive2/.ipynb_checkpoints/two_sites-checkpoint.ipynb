```python
def Dxyz1(N, α, λ, E_n, cutoff):
    
    # define annihilation operator according to Fock space
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    
    # displacement operator on cavity 1 which there are only two sites 
    if N == 2:
        cavity_subsystem = tensor(α*a.dag() - np.conj(α)*a)
        qubit_subsystem = tensor(σ_x, σ_x) + tensor(σ_y, σ_y) + λ*tensor(σ_z, σ_z) - E_n*tensor(I2, I2)
        
        return (tensor(cavity_subsystem, qubit_subsystem)).expm()
        
    # displacement operator on cavity 2 which there are three sites sites
    elif N == 3: 
        
        # define cavity and qubit subsystems for the first bond  
        cavity_subsystem1 = tensor(α*a.dag() - np.conj(α)*a, Ib)
        qubit_subsystem1 = tensor(σ_x, σ_x, I2) + tensor(σ_y, σ_y, I2) + λ*tensor(σ_z, σ_z, I2) - E_n*tensor(I2, I2, I2)
        
        # define cavity and qubit subsystems for the second bond  
        cavity_subsystem2 = tensor(Ib, α*a.dag() - np.conj(α)*a)
        qubit_subsystem2 = tensor(I2, σ_x, σ_x) + tensor(I2, σ_y, σ_y) + λ*tensor(I2, σ_z, σ_z) - E_n*tensor(I2, I2, I2)
        
    return [(tensor(cavity_subsystem1, qubit_subsystem1)).expm(), (tensor(cavity_subsystem2, qubit_subsystem2)).expm()]
def Dxyz2(N, α, λ, E_n, cutoff):
    
    # define annihilation operator according to Fock space
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    
    # displacement operator on cavity 1 which there are only two sites 
    if N == 2:
        cavity_subsystem = tensor([α*a.dag() - np.conj(α)*a, α*a.dag() - np.conj(α)*a])
        qubit_subsystem = tensor(σ_x, σ_x) + tensor(σ_y, σ_y) + λ*tensor(σ_z, σ_z) - E_n*tensor(I2, I2)
        
        return (tensor(cavity_subsystem, qubit_subsystem)).expm()
def ground_state_projection(N, cutoff, initial_state_list, num_states, λ, E_n, α_start, α_end, α_steps, singlet = True):

    # define operators for bosonic modes 
    a = destroy(cutoff)
    Ib = identity(cutoff)

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # define array over which we will sweep α
    α_array = np.linspace(α_start, α_end, α_steps)

    # define a singlet and density matrix for a singlet state
    S = tensor(vacuum, (tensor(up, down) - tensor(down, up)).unit())

    # define projection operator that projects the first cavity onto the vacuum
    P = tensor(vacuum*vacuum.dag(), I2, I2)

    # calcuate overlap with the singlet after displacement
    projection_fidelity_list = [[fidelity(ρ_ground_state2, \
                                        ((P*Dxyz1(N, α, λ, E_n, cutoff)*initial_state).unit()).ptrace([1,2]))**2 \
                                 for α in α_array] for initial_state in initial_state_list]

    # create plot 
    fig, ax = plt.subplots(figsize=(12, 9))

    # graph the projection of all initial states onto the singlet 
    for projection in projection_fidelity_list: 
        ax.plot(α_array, projection, label=f'Initial Fidelity =  {projection[0] : .3f}')


    #     for vacuum_prob in vacuum_prob_list:
    #         ax.plot(α_array, vacuum_prob, label=f'P(Vacuum), (Asymptote = {vacuum_prob[-1] : .3f} )')

    #     for element in zip(projection_fidelity_list, vacuum_prob_list):
    #         ax.plot(α_array, np.multiply(element[0], element[1]), \
    #                 label=f'P(Success) = P(Success | Vacuum) P(Vacuum) = {np.mean(np.multiply(element[0], element[1])) : .3f}')

    # plot ground state projection overlap with singlet as initial state
    # if singlet == True:
    #     initial_state_singlet = (tensor(vacuum, up, down) - tensor(vacuum, down, up)).unit()

    #     singlet_projection = [fidelity(ρ_ground_state, \
    #                                  ((P*Dxyz(N, α, λ, E_n, cutoff)*initial_state_singlet).unit()).ptrace([1,2])) \
    #                           for α in α_array]

    #     # plot projection onto singlet with singlet as initial state
    #     ax.plot(α_array, singlet_projection, label = 'Singlet')

    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('Fidelity', fontsize=20)
    ax.set_title(f'Simulated Projection Fidelity with N = {N}', fontsize=24)
    ax.legend(loc = 'center right')
    ax.grid(True)
def ground_state_projection2(N, cutoff, initial_state_list, num_states, λ, E_n, α_start, α_end, α_steps, singlet = True):

    # define operators for bosonic modes 
    a = destroy(cutoff)
    Ib = identity(cutoff)

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # define array over which we will sweep α
    α_array = np.linspace(α_start, α_end, α_steps)

    # define a singlet and density matrix for a singlet state
    S = tensor(vacuum, (tensor(up, down) - tensor(down, up)).unit())

    # define projection operators that projects the first and second cavities onto the vacuum
    P1 = tensor(vacuum*vacuum.dag(), I2, I2)
    P2 = tensor(vacuum*vacuum.dag(), vacuum*vacuum.dag(), I2, I2)
    
    # calcuate overlap with the singlet after displacement
    fidelity_array_list1 = [[fidelity(ρ_ground_state2, \
                                        ((P1*Dxyz1(N, α, λ, E_n, cutoff)*initial_state).unit()).ptrace([1,2]))**2 \
                                 for α in α_array] for initial_state in initial_state_list[0]]

    # calcuate overlap with the singlet after displacement
    fidelity_array_list2 = [[fidelity(ρ_ground_state2, \
                                        ((P2*Dxyz2(N, α, λ, E_n, cutoff)*initial_state).unit()).ptrace([2,3]))**2 \
                                 for α in α_array] for initial_state in initial_state_list[1]]
    
    # define list of colors to use for plots 
    colors = ['blue', 'orange', 'green','red', 'purple', 'black', 'pink', 'gray', 'cyan', 'magenta']

    # create plot 
    fig, ax = plt.subplots(figsize=(12, 9))

    # graph the theoretical projection of all initial states onto the true ground state 
    for i, fidelity_array in enumerate(fidelity_array_list1): 
        ax.plot(α_array, fidelity_array, label=f'Projecting 1 Cavity, (Initial Fidelity = {fidelity_array[0] : .3f})', \
                color=colors[i], linestyle='--')

    # graph the theoretical projection of all initial states onto the true ground state 
    for i, fidelity_array in enumerate(fidelity_array_list2): 
        ax.plot(α_array, fidelity_array, label=f'Projecting 2 Cavities, (Initial Fidelity = {fidelity_array[0] : .3f})', \
                color=colors[i], linestyle='-')


    #     for vacuum_prob in vacuum_prob_list:
    #         ax.plot(α_array, vacuum_prob, label=f'P(Vacuum), (Asymptote = {vacuum_prob[-1] : .3f} )')

    #     for element in zip(projection_fidelity_list, vacuum_prob_list):
    #         ax.plot(α_array, np.multiply(element[0], element[1]), \
    #                 label=f'P(Success) = P(Success | Vacuum) P(Vacuum) = {np.mean(np.multiply(element[0], element[1])) : .3f}')

    # plot ground state projection overlap with singlet as initial state
    # if singlet == True:
    #     initial_state_singlet = (tensor(vacuum, up, down) - tensor(vacuum, down, up)).unit()

    #     singlet_projection = [fidelity(ρ_ground_state, \
    #                                  ((P*Dxyz(N, α, λ, E_n, cutoff)*initial_state_singlet).unit()).ptrace([1,2])) \
    #                           for α in α_array]

    #     # plot projection onto singlet with singlet as initial state
    #     ax.plot(α_array, singlet_projection, label = 'Singlet')

    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('Fidelity', fontsize=20)
    ax.set_title(f'Simulated Projection Fidelity with N = {N}', fontsize=24)
    ax.legend(loc = 'center right')
    ax.grid(True)
```

```python
def plot_theoretical_fidelity(N, initial_state_list, cutoff, α_start, α_end, α_steps):

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # define singlet and triplet states with cavity in vacuum
    S = tensor(vacuum, (tensor(up, down) - tensor(down, up)).unit())
    T1 = tensor(vacuum, up, up)
    T0 = tensor(vacuum, (tensor(up, down) + tensor(down, up)).unit())
    Tn1 = tensor(vacuum, down, down)

    # create array over which to sweep α
    α_array = np.linspace(α_start, α_end, α_steps)

    # collect initial state and probability amplitudes in the singlet-triplet basis 
    # where these amplitudes correspond to |s_0|^2, |t_1|^2, |t_0|^2, and |t_{-1}|^2
    # collect a list of these probability amplitudes for each initial stat
    mod_sq_s0_list = [fidelity(state, S)**2 for state in initial_state_list] 
    mod_sq_t1_list = [fidelity(state, T1)**2 for state in initial_state_list]
    mod_sq_t0_list = [fidelity(state, T0)**2 for state in initial_state_list]
    mod_sq_tn1_list = [fidelity(state, Tn1)**2 for state in initial_state_list]

    # for each value of α plot the theoretical fidelity for each state 
    fidelity_theory_list = [[element[0]/(element[0] + np.exp(-16*α**2)*(element[1] + element[2] + element[3])) \
                             for α in α_array] \
                             for element in zip(mod_sq_s0_list, mod_sq_t1_list, mod_sq_t0_list, mod_sq_tn1_list)] 

    # create plot with axes
    fig, ax = plt.subplots(figsize=(12, 9))

    # graph the theoretical projection of all initial states onto the singlet 
    for fidelity_array in fidelity_theory_list: 
        ax.plot(α_array, fidelity_array, label=f'P(Success | Vacuum), (Initial Fidelity = {fidelity_array[0] : .3f})')
    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('Fidelity', fontsize=20)
    ax.set_title(f'Manually Defined Projection Fidelity with N = {N}', fontsize=24) 
    ax.legend(fontsize = 15)
    ax.grid(True)
```
