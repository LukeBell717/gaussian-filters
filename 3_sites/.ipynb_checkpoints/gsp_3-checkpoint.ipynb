# N-site Spin Chain

$$ \Large  H_N =  \sum_{j=1}^{N-1} \left( J_x X_j X_{j+1} + J_y Y_j Y_{j+1} + J_z Z_j Z_{j+1}\right) $$

```python
def spin_chain_quantities(N, Jx, Jy, Jz): 
    
    # Interaction coefficients, which we assume are uniform throughout the lattice
    Jx_list = Jx*np.ones(N)
    Jy_list = Jy*np.ones(N)
    Jz_list = Jz*np.ones(N)

    # Setup operators for individual qubits; 
    # here sx_list[j] = X_j, sy_list[j] = Y_j, and sz_list[j] = Z_j
    # since the Pauli matrix occupies the jth location in the tensor product of N terms
    # of which N-1 terms are the identity
    sx_list, sy_list, sz_list = [], [], []

    for i in range(N):
        op_list = [qeye(2)]*N
        op_list[i] = σ_x
        sx_list.append(tensor(op_list))
        op_list[i] = σ_y
        sy_list.append(tensor(op_list))
        op_list[i] = σ_z
        sz_list.append(tensor(op_list))

    # generate Hamiltonian by computing XX, YY and ZZ interactions in Hamiltonian
    H = 0 
    for j in range(N - 1):
        H += Jx_list[j] * sx_list[j] * sx_list[j + 1]
        H += Jy_list[j] * sy_list[j] * sy_list[j + 1]
        H += Jz_list[j] * sz_list[j] * sz_list[j + 1]

    # find eigenavlues and eigenstates of Hamiltonian 
    eigenvalues, eigenstates = H.eigenstates()

    # find indices of smallest eigenvalues, which correspond to the energy(ies) 
    # of the ground state (space in the case of degeneracy); 
    min_eigenvalue = min(eigenvalues)
    indices = [index for index, value in enumerate(eigenvalues) if value == min_eigenvalue]

    # find eigenstates corresponding to ground state
    eigenstates_list = eigenstates[indices]

    # create sum of density matrices of ground states in case ground state is degenerate
    ρ_ground_state = 0 
    for j in range(len(eigenstates_list)):
        ρ_ground_state += (eigenstates_list[j])*(eigenstates_list[j]).dag()

    # return normalized ground state
    return H, eigenstates, min_eigenvalue, ρ_ground_state 
def verify_orthonormality(eigenstates, tolerance):
    # find the inner product between all eigenstates; 
    # note that if eigenstates are orthonormal this should form 
    # the identity matrix with dimension equal to number of eigenstates
    fidelity_list = []
    for i, eigenstate in enumerate(eigenstates):
        row = []
        for j in range(len(eigenstates)):
            row.append(fidelity(eigenstate, eigenstates[j]))
        fidelity_list.append(np.array(row))


    # round entry in fidelity_matrix according to value of tolerance
    # very whether this matrix if equal to the identity
    if np.array_equal(np.round(fidelity_list, decimals=abs(int(np.log10(tolerance)))), np.eye(len(eigenstates))): 
        print('Your eigenstates form an orthonormal basis!')
    else: 
        print('Your eigenstates are not orthonormal')
```

# Operators 

$$ \Large D(\alpha) = e^{\alpha a^{\dagger} - \alpha^{*} a}$$

```python
# displacement operator 
def D(α, cutoff):
    
    # define operator for bosonic modes 
    a = destroy(cutoff)
    
    return (α*a.dag() - np.conj(α)*a).expm()
```

$$ \Large \text{BS}^{(i,j)}(\theta, \varphi) = e^{-i \frac{\theta}{2} \left[e^{i \varphi}a^{\dagger}_i a_j + e^{-i \varphi} a_i  a^{\dagger}_j  \right]} $$

```python
# beamsplitter gate 
def BS(θ,ϕ, cutoff):
    
    a = destroy(cutoff)
    
    # cavity_1 ⊗ cavity_2 ⊗ qubit_1⊗ qubit_2
    term = tensor([a.dag(), a, I2, I2, I2])
    phase = np.exp(1j*ϕ) 
    return (-1j*(θ/2)*(phase*term + np.conj(phase)*term.dag())).expm()
```

$$ \Large D_{XYZ}(\alpha, \lambda, E_n) = e^{ \left(\alpha a^{\dagger} - \alpha^{*} a \right) \otimes \left( X_1 X_2 + Y_1 Y_2 + \lambda Z_1 Z_2 - E_n \right)} $$

```python
# define D_{XYZ}
def Dxyz(N, α, λ, E_n, cutoff):
    
    # define annihilation operator according to Fock space
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    
    # displacement operator on cavity 1 which there are only two sites 
    if N == 2:
        cavity_subsystem = tensor(α*a.dag() - np.conj(α)*a)
        qubit_subsystem = tensor(σ_x, σ_x) + tensor(σ_y, σ_y) + λ*tensor(σ_z, σ_z) - E_n*tensor(I2, I2)
        
        return (tensor(cavity_subsystem, qubit_subsystem)).expm()
        
    # displacement operator on cavity 2 which there are three sites sites
    elif N == 3: 
        
        # define cavity and qubit subsystems for the first bond  
        cavity_subsystem1 = tensor(α*a.dag() - np.conj(α)*a, Ib)
        qubit_subsystem1 = tensor(σ_x, σ_x, I2) + tensor(σ_y, σ_y, I2) + λ*tensor(σ_z, σ_z, I2) - E_n*tensor(I2, I2, I2)
        
        # define cavity and qubit subsystems for the second bond  
        cavity_subsystem2 = tensor(Ib, α*a.dag() - np.conj(α)*a)
        qubit_subsystem2 = tensor(I2, σ_x, σ_x) + tensor(I2, σ_y, σ_y) + λ*tensor(I2, σ_z, σ_z) - E_n*tensor(I2, I2, I2)
        
    return [(tensor(cavity_subsystem1, qubit_subsystem1)).expm(), (tensor(cavity_subsystem2, qubit_subsystem2)).expm()]
```

# Ground State Projection for N = 2

```python
def generate_initial_states(cutoff, num_states):

    # define vacuum state for a cavity
    vacuum = basis(cutoff, 0)

    # create list of initial states
    return [(tensor(vacuum, rand_ket(2), rand_ket(2))).unit() for state in range(num_states)]
def ground_state_projection2(N, cutoff, initial_state_list, num_states, λ, E_n, α_start, α_end, α_steps, singlet = True):

    # define operators for bosonic modes 
    a = destroy(cutoff)
    Ib = identity(cutoff)

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # define array over which we will sweep α
    α_array = np.linspace(α_start, α_end, α_steps)

    # define a singlet and density matrix for a singlet state
    S = tensor(vacuum, (tensor(up, down) - tensor(down, up)).unit())
    ρ_ground_state = ρ_ground_state_2

    # define projection operator that projects the first cavity onto the vacuum
    P = tensor(vacuum*vacuum.dag(), I2, I2)

    # calcuate overlap with the singlet after displacement
    projection_fidelity_list = [[fidelity(ρ_ground_state, \
                                        ((P*Dxyz(N, α, λ, E_n, cutoff)*initial_state).unit()).ptrace([1,2]))**2 \
                                 for α in α_array] for initial_state in initial_state_list]

    # create plot 
    fig, ax = plt.subplots(figsize=(12, 9))

    # graph the projection of all initial states onto the singlet 
    for projection in projection_fidelity_list: 
        ax.plot(α_array, projection, label=f'P(Success | Vacuum) , (Initial Fidelity =  {projection[0] : .3f} )')


    #     for vacuum_prob in vacuum_prob_list:
    #         ax.plot(α_array, vacuum_prob, label=f'P(Vacuum), (Asymptote = {vacuum_prob[-1] : .3f} )')

    #     for element in zip(projection_fidelity_list, vacuum_prob_list):
    #         ax.plot(α_array, np.multiply(element[0], element[1]), \
    #                 label=f'P(Success) = P(Success | Vacuum) P(Vacuum) = {np.mean(np.multiply(element[0], element[1])) : .3f}')

    # plot ground state projection overlap with singlet as initial state
    # if singlet == True:
    #     initial_state_singlet = (tensor(vacuum, up, down) - tensor(vacuum, down, up)).unit()

    #     singlet_projection = [fidelity(ρ_ground_state, \
    #                                  ((P*Dxyz(N, α, λ, E_n, cutoff)*initial_state_singlet).unit()).ptrace([1,2])) \
    #                           for α in α_array]

    #     # plot projection onto singlet with singlet as initial state
    #     ax.plot(α_array, singlet_projection, label = 'Singlet')

    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('Fidelity', fontsize=20)
    ax.set_title('Singlet Projection Fidelity after Circuit', fontsize=24)
    ax.legend(loc = 'center right')
    ax.grid(True)
def plot_theoretical_fidelity(initial_state_list, cutoff, α_start, α_end, α_steps):

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # define singlet and triplet states with cavity in vacuum
    S = tensor(vacuum, (tensor(up, down) - tensor(down, up)).unit())
    T1 = tensor(vacuum, up, up)
    T0 = tensor(vacuum, (tensor(up, down) + tensor(down, up)).unit())
    Tn1 = tensor(vacuum, down, down)

    # create array over which to sweep α
    α_array = np.linspace(α_start, α_end, α_steps)

    # collect initial state and probability amplitudes in the singlet-triplet basis 
    # where these amplitudes correspond to |s_0|^2, |t_1|^2, |t_0|^2, and |t_{-1}|^2
    # collect a list of these probability amplitudes for each initial stat
    mod_sq_s0_list = [fidelity(state, S)**2 for state in initial_state_list] 
    mod_sq_t1_list = [fidelity(state, T1)**2 for state in initial_state_list]
    mod_sq_t0_list = [fidelity(state, T0)**2 for state in initial_state_list]
    mod_sq_tn1_list = [fidelity(state, Tn1)**2 for state in initial_state_list]

    # for each value of α plot the theoretical fidelity for each state 
    fidelity_theory_list = [[element[0]/(element[0] + np.exp(-16*α**2)*(element[1] + element[2] + element[3])) \
                             for α in α_array] \
                             for element in zip(mod_sq_s0_list, mod_sq_t1_list, mod_sq_t0_list, mod_sq_tn1_list)] 

    # create plot with axes
    fig, ax = plt.subplots(figsize=(12, 9))

    # graph the theoretical projection of all initial states onto the singlet 
    for fidelity_array in fidelity_theory_list: 
        ax.plot(α_array, fidelity_array, label=f'Initial Fidelity = {fidelity_array[0] : .3f}')
    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('Fidelity', fontsize=20)
    ax.set_title('Theoretical Singlet Projection Fidelity', fontsize=24) 
    ax.legend(fontsize = 20)
    ax.grid(True)
```

```python
N = 2
cutoff = 20
num_states = 5
cavity = 1
λ = 1
E_n = -3
α_start = 0
α_end = 1
α_steps = 100
initial_state_list = generate_initial_states(cutoff, num_states)
ground_state_projection2(N, cutoff, initial_state_list, num_states, λ, E_n, α_start, α_end, α_steps, singlet = True)
plot_theoretical_fidelity(initial_state_list, cutoff, α_start, α_end, α_steps)
```

# Convergence Analysis for Trotterization where  N = 3

$$ \Large \forall \epsilon > 0, \ \exists R \in \mathbb{N} \text{ such that } \| D_{\text{exact}}(\alpha) - D_{\text{Trotter}}(\alpha, r) \|_{F} < \epsilon \text{ for } r \ge R $$ 

$$ \Large \left( \text{where } \| \cdot \|_{F} \text{ is the Frobenius norm } \right)$$

```python
def convergence_analysis(D_exact, D_approx, α, ϵ):
    
    # compute Frobenius norm for one order of Trotterization
    r = 1
    frobenius_norm = np.linalg.norm(D_exact(α) - D_approx(α, r), 'fro')

    # find value of r that satisfies || D_exact - D_approx ||_F
    while frobenius_norm >= ϵ:
        r += 1
        frobenius_norm = np.linalg.norm(D_exact(α) - D_approx(α, r), 'fro')
    print(f'For r = {r} and \u03B1 = {α} the Frobenius norm is {frobenius_norm:.4f} ')
```

$$ \Large D^{(j)}_{\text{{k}}}(\alpha, \lambda, E_n) \equiv e^{ \left(\alpha a^{\dagger}_j - \alpha^{*} a_j \right) \otimes \left( H_\text{{k}} - E_n \right)} $$

```python
# operator that displaces cavity 1 by energy between spins 1,2, and 3
def D1_tot(α, E_n, cutoff):
    
    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(α*a.dag() - np.conj(α)*a, Ib, (H3 - E_n*tensor(I2, I2,I2)))).expm() 

# operator that displaces cavity 2 by energy between spins 1,2, and 3
def D2_tot(α, E_n, cutoff):
    
    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(Ib, α*a.dag() - np.conj(α)*a, (H3 - E_n*tensor(I2, I2,I2)))).expm() 

# operator that displaces cavity 1 by energy between spins 1 and 2
def D1_12(α, cutoff):

    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(α*a.dag() - np.conj(α)*a, Ib, tensor(H2,I2))).expm() 

# operator that displaces cavity 1 by energy between spins 2 and 3
def D1_23(α, cutoff):

    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(α*a.dag() - np.conj(α)*a, Ib, tensor(I2,H2))).expm()

# operator that displaces cavity 2 by energy between spins 1 and 2
def D2_12(α, cutoff):

    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(Ib, tensor(α*a.dag() - np.conj(α)*a, tensor(H2,I2)))).expm() 

# operator that displaces cavity 2 by energy between spins 2 and 3
def D2_23(α, cutoff):

    # define bosonic operators 
    a = destroy(cutoff)
    Ib = qeye(cutoff)
    return (tensor(Ib, tensor(α*a.dag() - np.conj(α)*a, tensor(I2,H2)))).expm()

# operator that uses Trotterization to displace cavity 1 by energy between spins 1,2 and 3
def D1_tot_trotter(α, E_n, r, cutoff):
    
    # define bosonic operators 
    Ib = qeye(cutoff)
    
    # return Trotterized conditional displacement operator that 
    # that is unconditionally shifted by α*E_n
    return ((D1_23(α/r, cutoff)*D1_12(α/r, cutoff))**r)*tensor(D(-E_n*α, cutoff),Ib,I2,I2,I2)

# operator that uses Trotterization to displace cavity 2 by energy between spins 1,2 and 3
def D2_tot_trotter(α, E_n, r, cutoff):
    
    # define bosonic operators 
    Ib = qeye(cutoff)
    
    # return Trotterized conditional displacement operator that 
    # that is unconditionally shifted by α*E_n
    return ((D2_23(α/r, cutoff)*D2_12(α/r, cutoff))**r)*tensor(Ib,D(-E_n*α, cutoff),I2,I2,I2)

# operator that uses parallelized Trotter
def D12_tot_trotter(α, E_n, r, cutoff):
    
    # define bosonic operators 
    Ib = qeye(cutoff)
    
    # return Trotterized conditional displacement operator that 
    # that is unconditionally shifted by α*E_n
    return (D2_12(α/r, cutoff)*D1_23(α/r, cutoff)*(D2_23(α/r, cutoff)*D1_12(α/r, cutoff))**r) \
            *tensor(D(-E_n*α, cutoff),D(-E_n*α, cutoff),I2,I2,I2)
```

$$ \Large \mathcal{D}^{(1)}_{123}(\alpha, \lambda, E_n) \approx \left(\mathcal{D}^{(1)}_{23} \left( \alpha/r, \lambda, E_n \right) \mathcal{D}^{(1)}_{12}(\alpha/r, \lambda, E_n)\right)^r $$

$$ \Large e^{\left( \alpha a^{\dagger}_1 - \alpha^{*}a_1 \right) \left(H_{123} - E_n \right)} \approx \left( e^{\left( \frac{\alpha}{r} a^{\dagger}_1 - \frac{\alpha^{*}}{r} a_1 \right) \left(H_{23} - E_n \right) } e^{\left( \frac{\alpha}{r} a^{\dagger}_{1} - \frac{\alpha^{*}}{r} a_{1} \right) \left(H_{12} - E_n \right)} \right)^r $$ 

```python
# define parameters
α = 1
ϵ = 1
E_n = min_eigenvalue_3
cutoff = 5

# define functions we wish to compare
GSP1_tot = lambda α: D1_tot(α, E_n, cutoff)
GSP1_trotter = lambda α, r: D1_tot_trotter(α, E_n, r, cutoff)

convergence_analysis(GSP1_tot, GSP1_trotter, α, ϵ)
```

$$ \Large \mathcal{D}^{(2)}_{123}(\alpha, \lambda, E_n) \approx \left( \mathcal{D}^{(2)}_{12}(\alpha/r, \lambda, E_n) \mathcal{D}^{(2)}_{23}(\alpha/r, \lambda, E_n)\right)^r $$

$$ \Large e^{\left( \alpha a^{\dagger}_2 - \alpha^{*}a_2 \right) \left(H_{123} - E_n \right)} \approx \left( e^{\left( \frac{\alpha}{r} a^{\dagger}_{2} - \frac{\alpha^{*}}{r} a_{2} \right) \left(H_{12} - E_n \right)} e^{\left( \frac{\alpha}{r} a^{\dagger}_2 - \frac{\alpha^{*}}{r} a_2 \right) \left(H_{23} - E_n \right)} \right)^r $$ 

```python
α = 1
ϵ = 1
GSP2_tot = lambda α: D2_tot(α, E_n, cutoff)
GSP2_trotter = lambda α, r: D2_tot_trotter(α, E_n, r, cutoff)

convergence_analysis(GSP2_tot, GSP2_trotter, α, ϵ)
```

$$ \mathcal{D}^{(2)}_{123}(\alpha/r, \lambda, E_n) \mathcal{D}^{(1)}_{123}(\alpha/r, \lambda, E_n) \approx  \left( \mathcal{D}^{(2)}_{12}(\alpha/r, \lambda, E_n) \mathcal{D}^{(2)}_{23}(\alpha/r, \lambda, E_n)\right)^r \left(\mathcal{D}^{(1)}_{23} \left( \alpha/r, \lambda, E_n \right) \mathcal{D}^{(1)}_{12}(\alpha/r, \lambda, E_n)\right)^r $$

```python
α = 1/np.sqrt(2)
ϵ = 1
GSP12_tot = lambda α: D2_tot(α, E_n, cutoff)*D1_tot(α, E_n, cutoff)
GSP12_trotter = lambda α, r: D2_tot_trotter(α, E_n, r, cutoff)*D1_tot_trotter(α, E_n, r, cutoff)

convergence_analysis(GSP12_tot, GSP12_trotter, α, ϵ)
```

# Ground State Projection for N = 3

```python
def generate_initial_states(num_states, cutoff):
    
    # define vacuum 
    vacuum = basis(cutoff, 0)

    # generate a random three-qubit state with both cavities in vacuum
    return [tensor(vacuum, vacuum, rand_ket(2), rand_ket(2), rand_ket(2)) for initial_state in range(num_states)]
def ground_state_projection3(cavities, parallelized, num_states, initial_state_list, λ, E_n, r, α_start, α_end, steps, cutoff):

    # define array over which we will sweep α
    α_array = np.linspace(α_start, α_end, steps)

    # define operators for bosonic modes 
    a = destroy(cutoff)
    Ib = identity(cutoff)

    # define vacuum state for a cavity, and spin up and spin down states for qubits
    vacuum = basis(cutoff, 0)
    up = basis(2,0)
    down = basis(2,1)

    # find the ground state for spin chain when N = 3
    H3, eigenstates, E_n, ρ_ground_state = spin_chain_quantities(3, 1, 1, 1)
    
    if cavities == 1: 
        
        # define projection operator that projects both cavities onto the vacuum
        vacuum_projector = vacuum*vacuum.dag()
        P_vac = tensor(vacuum_projector, Ib, I2, I2, I2)
            
        # create titles
        title_string_exact = 'Exact GSP with Single Cavity'
        title_string_trotter = f'Trotterized GSP with Single Cavity and r = {r}'

        # define ground state preparation algorithm using Trotterization
        GSP_trotter = lambda α: D1_tot_trotter(α, E_n, r, cutoff) 

        # define ground state preparation algorithm without Trotterization
        GSP_exact = lambda α: D1_tot(α, E_n, cutoff)

    elif cavities == 2: 
        
        if parallelized: 
        
            # create titles 
            title_string_exact = 'Exact GSP with Two Cavities'
            title_string_trotter = f'Parallelized and Trotterized GSP with Two Cavities and r = {r}'

            # define projection operator that projects both cavities onto the vacuum
            vacuum_projector = vacuum*vacuum.dag()
            P_vac = tensor(vacuum_projector, vacuum_projector, I2, I2, I2) 

            # define ground state preparation algorithm using Trotterization
            GSP_trotter = lambda α: D12_tot_trotter(α, E_n, r, cutoff)

            # define ground state preparation algorithm without Trotterization
            GSP_exact = lambda α: D1_tot(α, E_n, cutoff)*D2_tot(α, E_n, cutoff)
            
        else: 
            
             # create titles 
            title_string_exact = 'Exact GSP with Two Cavities'
            title_string_trotter = f'Trotterized GSP with Two Cavities and r = {r}'

            # define projection operator that projects both cavities onto the vacuum
            vacuum_projector = vacuum*vacuum.dag()
            P_vac = tensor(vacuum_projector, vacuum_projector, I2, I2, I2) 

            # define ground state preparation algorithm using Trotterization
            GSP_trotter = lambda α: D1_tot_trotter(α, E_n, r, cutoff)*D2_tot_trotter(α, E_n, r, cutoff)

            # define ground state preparation algorithm without Trotterization
            GSP_exact = lambda α: D1_tot(α, E_n, cutoff)*D2_tot(α, E_n, cutoff)
            
            
    # compute projection fidelity data without Trotterization
    projection_fidelities_exact = [[fidelity(ρ_ground_state, \
                                            ((P_vac*GSP_exact(α)*initial_state).unit()).ptrace([2,3,4])) for α in α_array] \
                                    for initial_state in initial_state_list]

    # compute projection fidelity data using Trotterization
    projection_fidelities_trotter = [[fidelity(ρ_ground_state, \
                                            ((P_vac*GSP_trotter(α)*initial_state).unit()).ptrace([2,3,4])) for α in α_array] \
                                    for initial_state in initial_state_list]

    # create plot 
    # fig, ax = plt.subplots(figsize=(8, 6))
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # graph fidelity of exact GSP with true ground state
    for i, projection in enumerate(projection_fidelities_exact): 
        ax1.plot(α_array, projection, label=f'Initial Fidelity {projection[0] : .3f}')
        
    ax1.set_xlabel(r'$\alpha$')
    ax1.set_ylabel('Fidelity')
    ax1.set_title(title_string_exact)
    ax1.legend()
    ax1.grid(True)
        
    # graph fidelity of trotterized GSP with true ground state
    for i, projection in enumerate(projection_fidelities_trotter): 
        ax2.plot(α_array, projection, label=f'Initial Fidelity {projection[0] : .3f}')
    
    ax2.set_xlabel(r'$\alpha$')
    ax2.set_ylabel('Fidelity')
    ax2.set_title(title_string_trotter)
    ax2.legend()
    ax2.grid(True)
```

```python
def trotter_error(D_exact, D_trotter, r_start, r_end, r_steps, α_start, α_end, α_steps):

    # define array for α and array for order of Trotterization
    r_array = np.array(range(r_start, r_end + 1, r_steps))
    α_array = np.linspace(α_start, α_end, α_steps)

    # compute accuracy of trotter approximation according to Frobenius (Hilbert-Schmidt) norm of 
    # ||D_exact - D_trotter||_F
    frobenius_norm_list = [[np.linalg.norm((D_exact(α) - D_trotter(α, r)).full(), 'fro') \
      for r in r_array] for α in α_array]
    
    # create plot 
    fig, ax = plt.subplots(figsize=(8, 6))
    α_legend = r'$\alpha$'
    
    # # graph the projection of all initial states onto the singlet 
    for norm, α in zip(frobenius_norm_list, α_array): 
        ax.plot(r_array, norm, label = f'{α_legend} = {α}')
        
    ax.set_xlabel('r')
    ax.set_ylabel('Hilbert-Schmidt Norm')
    ax.set_title('Trotterization Approximation of D_tot')
    ax.legend()
    ax.grid(True)
```
