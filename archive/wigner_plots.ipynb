```python
# import relevant packages 
from qutip import *
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# define cutoff for Fock space, the annihilation operator, identity operator for a mode 
# and the displacement parameter α
cutoff = 11
a = destroy(cutoff)
Ib = identity(cutoff)
α = 2/np.sqrt(2)

# define density matrices of a coherent state, thermal state, and Fock statel
ρ_coherent = coherent_dm(cutoff, α)
ρ_thermal = thermal_dm(cutoff, 2)
ρ_fock = fock_dm(cutoff, 2)
```

```python
# define dimensions over which to Wigner functions 
x = np.linspace(-5,5,200)
y = x

# create the Wigner plots of a coherent, thermal, and Fock state
W_coherent = wigner(ρ_coherent, x, y)
W_thermal = wigner(ρ_thermal, x, y)
W_fock = wigner(ρ_fock, x, y)

# plot the results
fig, axes = plt.subplots(1, 3, figsize=(12,3))

# plot for coherent state
cont0 = axes[0].contourf(x, y, W_coherent, 100)
lbl0 = axes[0].set_title("Coherent state")

# plot for thermal state
cont1 = axes[1].contourf(x, y, W_thermal, 100)
lbl1 = axes[1].set_title("Thermal state")

# plot for Fock State
cont0 = axes[2].contourf(x, y, W_fock, 100)
lbl2 = axes[2].set_title("Fock state")

# display plots
plt.show()
```

```python
# create Wigner function of a superposition of Fock states with equal probability amplitude 
psi = (basis(cutoff, 0) + basis(cutoff, 1)).unit()
x = np.linspace(-5, 5, 500)
y = x
W = wigner(psi, x, y)

# generate features of Wigner colormap 
wmap = wigner_cmap(W)  
nrm = mpl.colors.Normalize(-W.max(), W.max())

# plot Wigner function 
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# plot with standard colormap 
plt0 = axes[0].contourf(x, y, W, 100, cmap=cm.RdBu, norm=nrm)
axes[0].set_title("Standard Colormap");
cb0 = fig.colorbar(plt0, ax=axes[0])

# plot with Wigner colormap 
plt1 = axes[1].contourf(x, y, W, 100, cmap=wmap)  # Apply Wigner colormap
axes[1].set_title("Wigner Colormap");
cb1 = fig.colorbar(plt1, ax=axes[1])

# display plots
fig.tight_layout()
plt.show()
```
